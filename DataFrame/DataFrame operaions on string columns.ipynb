```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('../data/Fortune_500_2017.csv', usecols=['Rank','Title','Industry','Hqlocation','Sector'])
fortune.head()
```

```python
fortune['Sector'] = fortune['Sector'].str.upper()
fortune.head()
```

```python
search = ['comp', 'Comp', 'COMP']
fortune['Industry'].str.contains('|'.join(search))
```

```python
fortune[['City', 'State']] = fortune['Hqlocation'].str.split(", ",expand=True)
```

```python
fortune.head()
```

```python
def build_shortcut(row):
    industry = row['Industry']
    result = ''
    for i in industry.split(' '):
        result += i[0]
    return result
```

```python
build_shortcut({'Industry': 'Factory Under Newspaper'})
```

```python
fortune['IndustryShort'] = fortune.apply(build_shortcut, axis=1)
fortune.head()
```
