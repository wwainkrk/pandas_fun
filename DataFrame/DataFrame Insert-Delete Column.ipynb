```python
import pandas as pd
```

```python
fuel = pd.read_csv('../data/fuel.csv', low_memory=False, usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'])
fuel.head()
```

```python
fuel['Combined KPL'] = fuel['Combined MPG (FT1)'] * 1.609 / 3.78
fuel.head()
```

```python
fuel.insert(loc=5, column='Liters per km', value = 100 / fuel["Combined KPL"])
```
