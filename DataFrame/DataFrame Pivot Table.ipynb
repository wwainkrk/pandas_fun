```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../data/marathon_results_2016.csv', index_col='Bib', usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df.head()
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df.head()
```

```python
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
df.pivot_table(values=['TotalSeconds', 'HalfSeconds'], index='Age', columns='M/F')
```

```python
df.pivot_table(values=['TotalSeconds', 'HalfSeconds'], index='Age', columns='M/F').swaplevel(axis=1).sort_index(axis=1).head()
```

```python
df.pivot_table(values=['Country'], index='Age', aggfunc='count').head()
```

```python
df.pivot_table(values='M/F', index='Country', aggfunc='count').sort_index(ascending=False).head()
```

```python
df.pivot_table(values='TotalSeconds', index=['Country', 'State', 'City'], aggfunc='mean').sort_index(ascending=False).head()
```
