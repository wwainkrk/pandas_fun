```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('../data/Fortune_500_2017.csv', usecols=['Rank','Title','Employees','Profits','Assets'], index_col='Rank')
fortune.head()
```

```python
fortune['RankByEmployees'] = fortune['Employees'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune['RankByProfits'] = fortune['Profits'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune.nlargest(3, 'Profits')
```

```python
fortune.nsmallest(3, 'Profits')
```

```python
isEmployeesRankFirst10 = fortune['RankByEmployees'] <= 10
```

```python
isProfitRankFirst10 = fortune['RankByProfits'] <= 10
```

```python
fortune[isEmployeesRankFirst10 & isProfitRankFirst10].head()
```

```python
isEmployeesRankMore400 = fortune['RankByEmployees'] <= 400
```

```python
fortune[~isEmployeesRankMore400 & isProfitRankFirst10]
```
